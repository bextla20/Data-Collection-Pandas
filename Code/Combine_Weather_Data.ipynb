```python
# Import libraries

import pandas as pd
from google.colab import drive
import glob # Import the glob module
import os # Import the os module
# Import necessary library
from datetime import datetime
```

```python
# Connect Google Drive
drive.mount('/content/drive')

csv_files = glob.glob(os.path.join("/content/drive/MyDrive/BEY_Reserach/Input_CSV", "*.csv"))
dfs = [pd.read_csv(file) for file in csv_files]
df2 = dfs[0]

# Assuming 'dfs' is a list of pandas DataFrames
for i in range(len(dfs)):
    dfs[i]['system:time_start'] = pd.to_datetime(dfs[i]['system:time_start'])

for i in range(1, len(dfs)):
    df2 = pd.merge(df2, dfs[i], on="system:time_start", how="left", suffixes=("", f"_{i}"))
df2
```

```python
# Drop columns with all zero values
df2_new = df2.loc[:, (df2 != 0).any(axis=0)]
df2_new
```

```python
# Access dt attribute correctly using .dt accessor
df2_new['Date'] = df2_new['system:time_start'].dt.year

df2_new
```
