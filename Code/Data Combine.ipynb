```python
    import pandas as pd

    def GetSkipRowsHeadSubHead(filePath):
        """
        Give file path for csv file
        returns rows to skip in an array (get it before reading csv and use as an args)
        returns headerIndex as an array[main,sub] return array of index for which any of the values are alphabets
        Note:
        Works if only table headers have alphabets (Mostly are )and Values are numbers.
        """
        Wholefile = []
        
        with open(filePath,'r') as f:
                rows_to_skip = [] #Stores rows to skip
                headerIndex = [] #store index of headers
                num_of_head_val = [] # Store index and number of values in a header [(hIndex,hValue)]
                            
                for i, line in enumerate(f):
                    values = line.split(',')
                    if len(values)<=1:
                        rows_to_skip.append(i) #append rows with num of value 1 or less than one for skippping

                    if any(c.isalpha() for c in line):
                        headerIndex.append(i)  #If it is alphabets counts it as header row (only works if data are  numbers)
                        num_of_head_val.append((i,len(values))) # Gets [(hIndex,hValue)]
                        
                headerIndex = [x for x in headerIndex if x not in rows_to_skip] # remove Main heading by removing index that is in rows_to_skip
                num_of_head_val = [x for x in num_of_head_val if x[0] not in rows_to_skip] #Same as above

                second_values = [t[1] for t in num_of_head_val] # This list have count of values in a header
                are_equal = all(value == second_values[0] for value in second_values) # Check if all header have same no: of values if true gets final output
        print(are_equal)
        if are_equal:
                print(f"The [(hIndex,hVal)]: {num_of_head_val}")                
                print(f"skiprows: {rows_to_skip}")
                print(f"Headers in file are at: {headerIndex}")
                return rows_to_skip,headerIndex,num_of_head_val
        else:
            Wholefile = [] # stores all lines
            with open(filePath,'r') as f:
                Wholefile = f.readlines()

            maxval_tuple = (max(num_of_head_val, key=lambda x: x[1])) #gets header with max number of value as (hIndex,hVal)
            missHeader = [t for t in num_of_head_val if t[1] != maxval_tuple[1]] #All [(hIndex,hVal)] with missing header value

            #add the difference to max no: header at the end of the header line as '' (empty space)
            for mi in missHeader:
                n=maxval_tuple[1]-mi[1]   #Number of empty space to add (ie, difference)
                line = Wholefile[mi[0]].strip()  # Remove newline character
                line += ',' + ','.join([''] * n)  # Add n empty values to the line
                Wholefile[mi[0]] = line + '\n' 

            #Write the change to the file.
            with open(filePath,'w') as f:                                             
                   f.writelines(Wholefile)
        print(f"The [(hIndex,hVal)]: {num_of_head_val}")                
        print(f"skiprows: {rows_to_skip}")
        print(f"Headers in file are at: {headerIndex}")
        return rows_to_skip,headerIndex,num_of_head_val


    def Modified_read_csv(path_to_data):
        """
        Get headers and subheaders and merge them 
        Note:
        This use the GetSkipRowsHeadSubHead(path_to_data) 
        
        """

        skipRows, headerIndex,hIdxhVal = GetSkipRowsHeadSubHead(path_to_data)
        headerIndex = [h-headerIndex[0] for h in headerIndex]
        df = pd.read_csv(path_to_data,skiprows=skipRows, header=headerIndex)

        if (df.shape[0]==365) & (df.shape[1]==max(hIdxhVal, key=lambda x: x[1])[1]):
            print("ignore warning if there is any!!!")
            headcolumns = []
            for i in headerIndex:
                a = df.columns.get_level_values(i).to_series()
                b = a.mask(a.str.startswith('Unnamed')).ffill().fillna('')
                headcolumns.append(b)
            df.columns = headcolumns
            df.columns = df.columns.map(''.join)
            df = df.dropna(axis=1, how='all')
            return df
        else:
            print(f"error shape of dataframe is not as expeted, Expected:{(365,max(hIdxhVal)[1])} , Obtained: {df.shape}")
```

```python
# Load SoilClimate data
#path is the file location
def SoilClimate(path):
    df1 = pd.read_csv(path, header=[1, 2])
    a = df1.columns.get_level_values(0).to_series()
    b = a.mask(a.str.startswith('Unnamed')).ffill().fillna('')
    df1.columns = [b, df1.columns.get_level_values(1)]
    df1.columns = df1.columns.map(''.join)
    return df1
```

```python
# Merged SoilClimate with rest of data
def combine(path, save, soilclimate, fieldCrop, Graze, SoilN, year):
    all_dfs = [] # create list
    for one_filename in glob.glob(path):
        new_df = pd.read_csv(one_filename, header=1) # load csv and drop first row.
        all_dfs.append(new_df) # add datarames to list
    
    Day_Climate, Day_FieldManage, Day_SoilC, Day_SoilMicrobe, Day_SoilP, Day_SoilWater = all_dfs # extract and create multiple csv from list
    df = pd.concat(all_dfs, axis=1).fillna(0) # combine dataframes
    df = pd.concat((df, soilclimate, fieldCrop, Graze, SoilN), axis=1) # combine dataframes
    df['year'] = year
    print(df.shape)
    df.to_csv(save) # export to csv
    return df.head() # Display data
```
